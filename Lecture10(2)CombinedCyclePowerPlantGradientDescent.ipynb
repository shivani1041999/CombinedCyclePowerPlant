```python
import numpy as np
import pandas as pd
train_data = np.loadtxt("0000000000002419_training_ccpp_x_y_train.csv",delimiter = ",")
test_data = np.loadtxt("0000000000002419_test_ccpp_x_test.csv",delimiter = ",")
x_train = train_data.copy()
x_train = np.insert(x_train,4,1,axis = 1)
x_test = test_data.copy()
x_test = np.insert(x_test,4,1,axis = 1)
# pd.DataFrame(x_train)
```

```python
def step_gradient(data,learning_rate,m):
    m_slope = np.zeros(data.shape[1] - 1)
    N = len(data)
    for i in range(N):
        y = data[i,5]
        for j in range(data.shape[1] - 1):
            c = data[i,j]
            s = 0
            for k in range(data.shape[1]-1):
                x = data[i,k]
                s += m[k]*x
            m_slope[j] += (-2/N)*(y-s)*c
            
    new_m = m - learning_rate * m_slope
    return new_m

def gd(data,learning_rate,num_iterations):
    m = np.zeros(data.shape[1] - 1)
    for i in range(num_iterations):
        m = step_gradient(data,learning_rate,m)
        print(i,"COst : ",cost(data,m))
    return m
def cost(data, m):
    total_cost = 0
    M = len(data)
    for i in range(M):
        y = data[i, 5]
        k = 0
        for j in range(data.shape[1] - 1):
            x = data[i, j]
            k += m[j]*x
        total_cost += (1/M)*((y - k)**2)
    return total_cost
def run(data):
    learning_rate = 0.00000096
    num_iterations = 4000
    m = gd(data,learning_rate,num_iterations)
    print(m)
    return m 
def y_pred(test_data,m):
    return np.sum(m*test_data,axis = 1)

m = run(x_train)
y_test_pred = y_pred(x_test,m)

print(pd.DataFrame(y_test_pred))
np.savetxt("y_ccpp",y_test_pred)
```
